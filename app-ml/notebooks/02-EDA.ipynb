```python
# Energy Efficiency EDA Notebook
# ==============================

# 1. Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load dataset
df =  pd.read_csv("../../data/raw_data/csv/energy-efficiency-data.csv")

# 3. Basic info
print(df.head())
print(df.describe())

# 4. Histograms for all features
df.hist(figsize=(15, 10), bins=20, color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

# 5. Correlation heatmap
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# 6. Scatter plots of Y1 and Y2 vs each input feature
features = ["X1","X2","X3","X4","X5","X6","X7","X8"]
targets = ["Y1","Y2"]

for target in targets:
    for feature in features:
        plt.figure(figsize=(6,4))
        sns.scatterplot(x=df[feature], y=df[target], alpha=0.7)
        plt.title(f"{target} vs {feature}")
        plt.xlabel(feature)
        plt.ylabel(target)
        plt.show()

# 7. Boxplots for categorical features vs targets
categorical_features = ["X6","X8"]

for target in targets:
    for cat in categorical_features:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=df[cat], y=df[target], palette="Set2")
        plt.title(f"{target} vs {cat}")
        plt.xlabel(cat)
        plt.ylabel(target)
        plt.show()
```
